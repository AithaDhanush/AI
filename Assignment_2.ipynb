```python
import numpy as np
x=np.random.uniform(-5,5)
eta=0.1
itermax=2
iter=1
while iter<=itermax:
  slope=(4*x*x*x) + (6*x)
  step_length=-eta*slope
  x=x+step_length
  iter=iter+1
print("control variable=",round(x))
obj=(x*x*x*x) + (3*x*x) + 10
print("objective function=",obj)
```
